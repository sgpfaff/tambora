All forces are subclasses of `BaseForce`, which only has an abstract `force` method. The types of forces are:

* **`ConservativeForce`**: Conservative forces that have an associated potential. Has an extra `potential` method. They are also computed between steps during integration. `BaseForce`s are calculated after full steps.
  
* **Self-gravity Forces** (`FalcONGravity` and `DirectSummationGravity`): Subclass of `ConservativeForce` for self-gravity.

Forces can be added directly, producing a `CompositeForce` instance.



```python
from tambora.dynamics import ExternalGalpyPotential, FalcONGravity
from galpy.potential import PlummerPotential

pot1 = PlummerPotential(amp=1, b=1)
pot2 = PlummerPotential(amp=2, b=2)
galpyforce1 = ExternalGalpyPotential(pot1)
galpyforce2 = ExternalGalpyPotential(pot2)
galpyforces = galpyforce1 + galpyforce2
falcONselfgravity = FalcONGravity(eps=0.01)
```
